### 성능 지표 확인

```python
import pandas as pd

data=pd.read_csv("resize_metrics.csv")
data
```

## loss 그래프(Train vs Val)

```python
loss=data[["Train Loss","Val Loss"]]
loss
```

```python
import matplotlib.pyplot as plt

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# Train Loss와 Val Loss 각각 다른 스타일로 플롯
plt.plot(loss["Train Loss"], label="Train Loss", color="blue", linestyle="-", linewidth=2)
plt.plot(loss["Val Loss"], label="Val Loss", color="orange", linestyle="-", linewidth=2)

# 축과 타이틀 설정
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.title("Train vs Validation Loss", fontsize=14)

# 그리드 추가
plt.grid(True, linestyle="--", alpha=0.5)

# 범례 위치 설정
plt.legend(loc="upper right", fontsize=12)

# 그래프 표시
plt.show()
```

### mIoU 그래프(Train vs Val)

```python
mIoU=data[["Train mIoU","Val mIoU"]]
mIoU
```

```python
import matplotlib.pyplot as plt

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# Train Loss와 Val Loss 각각 다른 스타일로 플롯
plt.plot(mIoU["Train mIoU"], label="Train mIoU", color="blue", linestyle="-", linewidth=2)
plt.plot(mIoU["Val mIoU"], label="Val mIoU", color="orange", linestyle="-", linewidth=2)

# 축과 타이틀 설정
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("mIoU", fontsize=12)
plt.title("Train vs Validation mIoU", fontsize=14)

# 그리드 추가
plt.grid(True, linestyle="--", alpha=0.5)

# 범례 위치 설정
plt.legend(loc="lower right", fontsize=12)

# 그래프 표시
plt.show()
```
